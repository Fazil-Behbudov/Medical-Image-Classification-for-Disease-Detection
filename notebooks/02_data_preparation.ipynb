# Data Preparation and Normalization

**Goal:** Get our preprocessed data ready for the neural network and save it so we can easily access it in future steps.

### 1. Import Libraries

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import cv2
from tqdm import tqdm
```

### 2. Load Preprocessed Data



```python
# Load the saved numpy arrays
results_dir = '../results'
X_train = np.load(os.path.join(results_dir, 'X_train.npy'))
y_train = np.load(os.path.join(results_dir, 'y_train.npy'))
X_test = np.load(os.path.join(results_dir, 'X_test.npy'))
y_test = np.load(os.path.join(results_dir, 'y_test.npy'))

print("Loaded data from .npy files:")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

### 3.Normalize the Pixel Values
Scale the pixel values from 0-255 to 0.0-1.0

```python
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

### 4. Reshape the Images

The CNN model expects a 4D tensor as input, which includes a "channel" dimension. Since our images are grayscale, we need to add a channel dimension of 1. The required shape is `(number_of_images, height, width, 1)`.

```python
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

print("Shape after adding channel dimension:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

### 5. Encode Labels

Our labels are currently strings (e.g., "glioma"). We need to convert these categorical text labels into numerical format so our model can understand them. We will use `LabelEncoder` from scikit-learn to assign a unique integer to each class.

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

### 6. Split Data for Validation

Now we'll split our training data (`X_train`, `y_train`) into a new, smaller training set and a validation set. The model will be trained on the new training set, and its performance will be evaluated on the validation set at the end of each epoch. This helps us monitor for overfitting. A common split is 80% for training and 20% for validation.

```python
'''
During training, we want to check how well our model is performing on data it hasn't been trained on.
This helps us see if the model is actually learning or just memorizing the training data (a problem called overfitting).
'''
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

### 7. Save the Final Datasets

This is the final step for this notebook. We will save all our prepared arrays (`X_train`, `y_train`, `X_val`, `y_val`, `X_test`, `y_test`) to the `results/` directory. This allows us to easily load them in the next notebook for model training without re-running all these preparation steps.

```python
results_dir = '../results'
if not os.path.exists(results_dir):
    os.makedirs(results_dir)

# Save the arrays
np.save(os.path.join(results_dir, 'X_train.npy'), X_train)
np.save(os.path.join(results_dir, 'y_train.npy'), y_train)
np.save(os.path.join(results_dir, 'X_test.npy'), X_test)
np.save(os.path.join(results_dir, 'y_test.npy'), y_test)
np.save(os.path.join(results_dir, 'X_val.npy'), X_val)
np.save(os.path.join(results_dir, 'y_val.npy'), y_val)

print("Data preparation completed and saved successfully in the 'results' directory.")
```
